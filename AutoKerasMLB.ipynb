```python
import autokeras
from autokeras import StructuredDataClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import pandas as pd
```

```python
data = pd.read_csv('Current Stats and Game Features.csv') # 53.74% Accuracy
data = pd.read_csv('Scraping/Previous Stats and Games.csv') # 53.3% Accuracy
# Your own models are better

# Split data 
X = data.loc[:,'H  #Bat':]
y = data['Home Win']

# Train, test, valid

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
# X_test, X_valid, y_test, y_valid = train_test_split(X_test, y_test, test_size=0.5)
```

```python
%%time
ak = StructuredDataClassifier(objective='val_accuracy')
ak.fit(x=X_train, y=y_train, verbose=0)
loss, acc = ak.evaluate(X_test, y_test, verbose=0)
print('Loss: ',loss)
print('Acc: ',acc)
```
